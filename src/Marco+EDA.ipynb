```python
import pandas as pd
df = pd.read_json('dev_v2.0.json')
```

```python
df.head()
```

```python
df.iloc[2,1]
```

```python
df['passages2'] = df['passages'].astype('str')
```

```python
df_mini = df[1:4000]
```

```python
df_mini.tail()
```

```python
df_mini['passage_len'] = df_mini['passages2'].str.split().apply(len)
```

```python
df_mini['answers2'] = df['answers'].astype('str')
df_mini['answer_len'] = df_mini['answers2'].str.split().apply(len)
```

```python
df_mini[58:68]
```

```python
import numpy as np
print(np.average(df_mini['passage_len']))
print(np.average(df_mini['answer_len']))
```

```python
table = pd.pivot_table(df_mini,columns ='query_id', values = 'passage_len', aggfunc=np.mean)
#table2 =  table.div()                   
print (table)
```

```python
table_ans = pd.pivot_table(df_mini,columns ='query_id', values = 'answer_len', aggfunc=np.mean)
#table2 =  table.div()                   
```

```python
table = pd.DataFrame(table)
table = table.transpose()
table

table_ans = pd.DataFrame(table_ans)
table_ans = table_ans.transpose()
```

```python
table.index
```

```python
table['passage_len']
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(table.index, table['passage_len'])
plt.xlabel('Query ID')
plt.ylabel("Average Length of Passage")
plt.title("Average Passage Length by Query")
```

```python
table2 = pd.crosstab(index = table['passage_len'],  columns=['count'])
table2 = pd.DataFrame(table2)
#table2
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(table2.index, table2['count'])
plt.xlabel('Length of Passages')
plt.ylabel("Count of Queries")
plt.title("Distribution of Length of Passages")
```

```python
table2_ans = pd.crosstab(index = table_ans['answer_len'],  columns=['count'])
table2_ans = pd.DataFrame(table2_ans)
#table2
```

```python
plt.bar(table2_ans.index, table2_ans['count'])
plt.xlabel('Length of Answers')
plt.ylabel("Count of Queries")
plt.title("Distribution of Length of Answers")
```
